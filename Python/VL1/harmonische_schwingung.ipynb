# Harmonische Schwingung

Plot einer Sinusschwingung mit Periodendauer-Markierungen.

[![Open In Colab](https://colab.research.google.com/assets/colab-badge.svg)](https://colab.research.google.com/github/hoeverc/laermarmesysteme_medien/blob/main/Python/VL1/harmonische_schwingung.ipynb)

```python
import numpy as np
import matplotlib.pyplot as plt

T=3*2*np.pi
t=np.linspace(0, T, 1000)

fig, ax=plt.subplots(figsize=(10, 3))
ax.plot(t, np.sin(t), 'k', linewidth=2)
ax.plot([0, T], [0, 0], 'k')
ax.plot([2.5*np.pi, 2.5*np.pi], [0, 1.25], 'k')
ax.plot([4.5*np.pi, 4.5*np.pi], [0, 1.25], 'k')
ax.set_xlim(2.8, 18)
ax.set_ylim(-1.5, 1.5)
ax.axis('off')
plt.tight_layout()
plt.show()
```
