# Cosinus-Addition und Effektivwert-Addition

Interaktive Demonstration der Addition zweier Cosinus-Signale
mit einstellbaren Frequenzen und Phasendifferenz.

[![Open In Colab](https://colab.research.google.com/assets/colab-badge.svg)](https://colab.research.google.com/github/hoeverc/laermarmesysteme_medien/blob/main/Python/VL2/CosinusAdditionEffektivwertAddition.ipynb)

```python
%matplotlib widget
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display

fontsize=14
linewidth=2
T=7*np.pi
t=np.linspace(0, T, 10000)
dt=t[1]-t[0]

fig, ax=plt.subplots(figsize=(10, 5))
plt.tight_layout()
plt.show()

def update_plot(freq1, freq2, phi_deg):
  """Plot der Cosinus-Addition."""
  phi=np.deg2rad(phi_deg)
  omega=2*np.pi*np.array([freq1, freq2])

  # Signal: cos((w1-w2)*t+phi)
  #       + cos((w1+w2)*t+phi)
  x=(np.cos((omega[0]-omega[1])*t+phi)
    +np.cos((omega[0]+omega[1])*t+phi))

  # Positive/negative Anteile trennen
  x_pos=x.copy()
  x_pos[x<0]=np.nan
  x_neg=x.copy()
  x_neg[x>=0]=np.nan
  sum_pos=np.sum(x[x>=0])*dt
  sum_neg=np.sum(x[x<0])*dt

  ax.cla()
  ax.plot(t, x_pos, linewidth=linewidth,
    label=f'x >= 0, sum = {sum_pos:.2f}')
  ax.plot(t, x_neg, linewidth=linewidth,
    label=f'x < 0, sum = {sum_neg:.2f}')
  ax.plot(t, np.zeros_like(t), 'k',
    linewidth=linewidth)
  ax.set_xlim(0, T)
  ax.set_xlabel(r'$t$ in s', fontsize=fontsize)
  ax.set_ylabel(
    r'$x(t)$ in a.u.', fontsize=fontsize)
  ax.set_title(
    r'$\cos((\omega_1+\omega_2)t)'
    r' + \cos((\omega_1-\omega_2)t)$',
    fontsize=20)
  ax.legend(fontsize=fontsize, loc='upper right')
  ax.grid(True)
  ax.tick_params(labelsize=fontsize)
  fig.tight_layout()
  fig.canvas.draw()

w_freq1=widgets.FloatSlider(
  value=0.5, min=0.1, max=3.0, step=0.01,
  description='Freq. 1 (Hz):',
  continuous_update=False)
w_freq2=widgets.FloatSlider(
  value=1.5, min=0.1, max=3.0, step=0.01,
  description='Freq. 2 (Hz):',
  continuous_update=False)
w_phi=widgets.FloatSlider(
  value=0, min=-180, max=180, step=1,
  description='Phase (°):',
  continuous_update=False)

out=widgets.interactive_output(
  update_plot,
  {'freq1': w_freq1, 'freq2': w_freq2, 'phi_deg': w_phi})

display(widgets.HBox([w_freq1, w_freq2, w_phi]))
```
