# Weber-Fechner-Gesetz

Darstellung des Weber-Fechner-Gesetzes auf linearer und
logarithmischer Achse.

[![Open In Colab](https://colab.research.google.com/assets/colab-badge.svg)](https://colab.research.google.com/github/hoeverc/laermarmesysteme_medien/blob/main/Python/VL2/weber_fechner.ipynb)

```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams.update({
  'lines.linewidth': 2,
  'font.size': 14,
})

R_R0=np.arange(0, 21, dtype=float)

# log10(0) ergibt -inf (wird nicht geplottet)
with np.errstate(divide='ignore'):
  E=2.3*np.log10(R_R0)

fig, (ax1, ax2)=plt.subplots(1, 2, figsize=(12, 4))

ax1.plot(R_R0, E)
ax1.set_xlabel(r'$R/R_0$')
ax1.set_ylabel(r'$E/E_0$')
ax1.grid(True)

ax2.semilogx(R_R0, E)
ax2.set_xlabel(r'$R/R_0$')
ax2.set_ylabel(r'$E/E_0$')
ax2.grid(True)
ax2.set_xlim(1, 20)

plt.tight_layout()
plt.show()
```
