# Frequenzachse

Darstellung der Frequenzachse mit Infraschall-, Hör- und Ultraschall-Bereich.

[![Open In Colab](https://colab.research.google.com/assets/colab-badge.svg)](https://colab.research.google.com/github/hoeverc/laermarmesysteme_medien/blob/main/Python/VL2/FrequenzAchse.ipynb)

```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams.update({
  'lines.linewidth': 2,
  'font.size': 14,
})

fig, ax=plt.subplots(figsize=(10, 3))
ax.semilogx([16, 16], [0, 2], ':k')
ax.semilogx([16000, 16000], [0, 2], ':k')
ax.text(2, 1.5, 'Infraschall')
ax.text(200, 1.5, 'H\u00f6rbereich')
ax.text(17000, 1.5, 'Ultraschall')
ax.set_xlim(1, 1e5)
ax.set_yticks([])
ax.set_xlabel(r'$f$ in Hz')
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.tight_layout()
plt.show()
```
