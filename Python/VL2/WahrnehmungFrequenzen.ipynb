# Wahrnehmung von Frequenzen (Weber-Fechner)

Demonstration des Weber-Fechner-Gesetzes für die
Frequenzwahrnehmung: gleiche absolute Frequenzdifferenz
bei niedrigen vs. hohen Frequenzen.

[![Open In Colab](https://colab.research.google.com/assets/colab-badge.svg)](https://colab.research.google.com/github/hoeverc/laermarmesysteme_medien/blob/main/Python/VL2/WahrnehmungFrequenzen.ipynb)

```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import Audio, display

# ===========================================================
# Signalfrequenzen, Hz
# ===========================================================
df=10
f0=[150, 1500]
freqs=[
  f0[0],                    # 150 Hz
  f0[0]+df,                 # 160 Hz
  f0[1],                    # 1500 Hz
  f0[1]+df,                 # 1510 Hz
  f0[1]+f0[1]/f0[0]*df,    # 1600 Hz
]

# Abtastrate und Signaldauer
fs=2*8192  # Hz
T_sig=4    # s
n=T_sig*fs
t=np.arange(n)/fs

# Signale erzeugen
signals=[
  np.cos(2*np.pi*freq*t) for freq in freqs
]
```

## Audio: niedrige Frequenzen

Vergleich von 150 Hz und 160 Hz (Differenz: 10 Hz = 6.7 %).

```python
print(f'Sound 1: {freqs[0]:.0f} Hz')
display(Audio(data=signals[0], rate=fs))
print(f'Sound 2: {freqs[1]:.0f} Hz')
display(Audio(data=signals[1], rate=fs))
print(f'Sound 3: {freqs[1]:.0f} Hz')
display(Audio(data=signals[1], rate=fs))
```

## Audio: hohe Frequenzen

Vergleich von 1500 Hz, 1510 Hz (10 Hz = 0.67 %)
und 1600 Hz (100 Hz = 6.7 %).

```python
print(f'Sound 4: {freqs[2]:.0f} Hz')
display(Audio(data=signals[2], rate=fs))
print(f'Sound 5: {freqs[3]:.0f} Hz')
display(Audio(data=signals[3], rate=fs))
print(f'Sound 6: {freqs[4]:.0f} Hz')
display(Audio(data=signals[4], rate=fs))
```

## Zeitsignale

```python
fontsize=14
linewidth=2

fig, (ax1, ax2)=plt.subplots(
  2, 1, figsize=(10, 6))

# Niedrige Frequenzen (Sounds 1-3)
ax1.plot(t, signals[0], linewidth=linewidth)
ax1.plot(t, signals[1], linewidth=linewidth)
ax1.plot(t, signals[1], linewidth=linewidth)
ax1.set_xlim(0, 1/freqs[0])
ax1.set_xlabel('t in s', fontsize=fontsize)
ax1.set_ylabel('Amplitude', fontsize=fontsize)
ax1.tick_params(labelsize=fontsize)
ax1.legend(
  [f'Signal 1: {freqs[0]:.0f} Hz',
   f'Signal 2: {freqs[1]:.0f} Hz',
   f'Signal 3: {freqs[1]:.0f} Hz'],
  loc='center left',
  bbox_to_anchor=(1, 0.5),
  fontsize=fontsize)

# Hohe Frequenzen (Sounds 4-6)
ax2.plot(t, signals[2], linewidth=linewidth)
ax2.plot(t, signals[3], linewidth=linewidth)
ax2.plot(t, signals[4], linewidth=linewidth)
ax2.set_xlim(0, 1/freqs[0])
ax2.set_xlabel('t in s', fontsize=fontsize)
ax2.set_ylabel('Amplitude', fontsize=fontsize)
ax2.tick_params(labelsize=fontsize)
ax2.legend(
  [f'Signal 4: {freqs[2]:.0f} Hz',
   f'Signal 5: {freqs[3]:.0f} Hz',
   f'Signal 6: {freqs[4]:.0f} Hz'],
  loc='center left',
  bbox_to_anchor=(1, 0.5),
  fontsize=fontsize)

plt.tight_layout()
plt.show()
```
